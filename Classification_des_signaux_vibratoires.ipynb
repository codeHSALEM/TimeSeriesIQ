# Auto Encodeur

```python
from tensorflow.keras.models import Model
import tensorflow as tf
```

```python
input_shape1=x_train.shape[1]
```

```python
# Création de l'autoencodeur
class Signaux(Model):
  def __init__(self):
    super(Signaux, self).__init__()
    self.encoder = tf.keras.Sequential([

        tf.keras.layers.Dense(x_train.shape[1], activation="relu",input_shape=(input_shape1,)),
        tf.keras.layers.Dense(400, activation="relu"),
        tf.keras.layers.Dense(50, activation="relu"),
     ])

    self.decoder = tf.keras.Sequential([
        tf.keras.layers.Dense(400, activation="relu"),
        tf.keras.layers.Dense(x_train.shape[1], activation="relu")
      ])

  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded

autoencoder1 = Signaux()
```

```python
autoencoder1.compile(optimizer="adam",
                    loss="mse"
                    )
```

```python
history1=autoencoder1.fit(x_train,
                         y_train,
                         epochs=30,
                         batch_size=256,
                         validation_data=(x_valid,y_valid)
                         )
```

```python
plt.plot(history1.history["loss"], label="Training Loss")
plt.plot(history1.history["val_loss"], label="Validation Loss")
plt.ylim([0, 3])
plt.legend()
```
